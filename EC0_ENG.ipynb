```python
import matplotlib.pyplot as plt
import numpy as np


x_data = np.array([2018, 2022, 2027, 2032])
y_data = np.array([3347364, 3371954, 3416451, 3476438])

plt.plot(x_data, y_data, marker='o', linestyle='--', color='red')
plt.xlabel("Year")
plt.ylabel("Number of popullation (millions)")
plt.title("Population Growth Prediction")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def y1(x1):
  return 19931250 / x1

def y2(x2):
  return 19394316 / x2

# Create x-axis values from 1 to 50 (inclusive)
x1 = np.arange(1, 16)
x2 = np.arange(1, 51)  # Using arange to ensure 50 is included

# Calculate y-axis values for both functions
y1_values = y1(x1)
y2_values = y2(x2)

plt.figure(figsize=(10, 5))

# Create the plot
plt.plot(x1, y1_values, label='Traditional Road', color='red')
plt.plot(x2, y2_values, label='Plastic Road', color='green')

# Set x-axis ticks with a step size of 5
plt.xticks(range(min(x1), max(x2) + 1, 5))  # Adjust range if needed


# Add labels and title
plt.xlabel('Years from construction')
plt.ylabel('Albanian Lek')
plt.title('Construction Cost Over Time')

# Add legend
plt.legend()

# Avoid division by zero error by setting a minimum x-axis value (optional)
# plt.xlim(1, 50)  # Uncomment this line to exclude x = 0

# Show the plot

plt.grid(True)
plt.show()
```
